```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("gender.csv")
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.columns = [i.strip() for i in data.columns]
```

```python
data.describe()
```

```python
data.nunique()
```

```python
# handle columns 
numeric_cols = [i for i in data.columns if(data[i].dtype in ['int64' ,'float64'])]
cat_col = [i for i in data.columns if i not in numeric_cols]
for i in cat_col:
    data[i] = [c.strip() for c in data[i]]
```

```python
new_data = pd.get_dummies(data , drop_first = True)
X = new_data.drop(['Gender_male'] , axis = 1)
y = new_data['Gender_male']
```

```python
data.corr()
```

LOGISTIC REGRESSION

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train , y_train)
```

```python
y_pred = logreg.predict(x_test)
accuracy_score(y_test , y_pred)
```

Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB , GaussianNB

nb = GaussianNB()
nb.fit(x_train , y_train)
```

```python
y_pred = nb.predict(x_test)
accuracy_score(y_test , y_pred)
```

SUPPORT VECTOR MACHINE(SVM)

```python
from sklearn import svm
sv = svm.SVC(kernel='linear')
sv.fit(x_train, y_train)
Y_pred_svm = sv.predict(x_test)

Y_pred_svm.shape
```

```python
score_svm = round(accuracy_score(Y_pred_svm,y_test),2)
print(score_svm)
```
